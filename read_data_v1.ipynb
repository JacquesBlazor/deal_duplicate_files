```python
# 找出所有的磁碟機和根目錄的組合
root_folders = list(df['磁碟機根目錄'].unique())
root_folders.sort()
for index, folder_name in enumerate(root_folders):
    print(index, folder_name)
```

```python
to_be_deleted = df[df['磁碟機根目錄'] == root_folders[0]][['路徑', '檔案名稱']]
to_be_deleted
```

```python
import os
```

```python
# 刪除檔案
for each_row in to_be_deleted.itertuples():
    file_name = os.path.join(each_row.路徑, each_row.檔案名稱)
    try:
        os.remove(file_name)
    except Exception as e:
        print('error: ', e)
    else:
        print(file_name)
```

```python
# 檢查檔案是否都刪除了
check_counter = 0
exist_files = []
root_folders = list(df['磁碟機根目錄'].unique())
root_folders.sort()
for index, folder_name in enumerate(root_folders):
    #print(index, folder_name)
    verify_if_exists = df[df['磁碟機根目錄'] == root_folders[0]][['路徑', '檔案名稱']]
    for each_row in verify_if_exists.itertuples():
        file_name = os.path.join(each_row.路徑, each_row.檔案名稱)
        if os.path.exists(file_name):
            exist_files.append(file_name)
        else:
            check_counter +=1
            print('.', end='')
print(check_counter, end='')
for each_file in exist_files:
    print(each_file)
```
