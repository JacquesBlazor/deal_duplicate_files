```python
import pandas as pd
import csv
import os
```

```python
group_by_path = f'E:/.developing_env/duplicated_files/group_by/'
df_file_lists = [file for file in os.listdir(group_by_path) if os.path.isfile(os.path.join(group_by_path, file))]
df_file_lists
```

```python
by_column = '檔案資料夾'
results = []
with open('E:/.developing_env/duplicated_files/log.csv', 'w', newline='', encoding='utf-8-sig') as f:
    csv_writer = csv.writer(f)
    csv_writer.writerow(['檔案資料夾', '來源路徑', '比較路徑', '結果', '備註'])
    for df_file in df_file_lists:
        print('df_file:', df_file, end='')
        path_frames = {}
        path_names = []
        folder_names = []
        counts = {}
        file_as_df = pd.read_csv(os.path.join(group_by_path, df_file))
        by_file_folder = file_as_df.groupby(by_column)
        for folder_name, folder_frame in by_file_folder:
            by_path = folder_frame.groupby('路徑')
            counts[folder_name] = by_path.count()
            folder_names.append(folder_name)
            compare_box = compare_name = None
            for path_name, path_frame in by_path:
                print('.', end='')
                path_names.append(path_name)
                path_frames[path_name] = path_frame.sort_values(['路徑', '檔案名稱'])[['組別', '檔案類型', '檔案資料夾', '檔案名稱', '大小', '修改日期']].reset_index(drop=True)
                if compare_box is None:
                    compare_box = path_frames[path_name]
                    compare_name = path_name
                    continue
                try:
                    if (compare_box == path_frames[path_name]).any(axis=None):
                        csv_writer.writerow([folder_name, compare_name, path_name, '相同'])
                    else:
                        csv_writer.writerow([folder_name, compare_name, path_name, '相異'])
                except Exception as e:
                    csv_writer.writerow([folder_name, compare_name, path_name, e])
        print()
```
