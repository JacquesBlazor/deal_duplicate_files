```python
import pandas as pd
import csv
import os
import re
```

```python
source_path = f'E:/.developing_env/duplicated_files/group_by/'
df_file_lists = [file for file in os.listdir(source_path) if os.path.isfile(os.path.join(source_path, file))]
```

```python
pointer = 0
results = pd.DataFrame()
with open('E:/.developing_env/duplicated_files/log.csv', 'w', newline='', encoding='utf-8-sig') as f:
    csv_writer = csv.writer(f)
    csv_writer.writerow(['檔案資料夾', '來源路徑', '比較路徑', '次數', '來源長度', '比較長度', '結果', '備註'])
    for df_file in df_file_lists:
        print('df_file:', df_file, end='')
        path_frames = {}
        path_names = []
        folder_names = []
        counts = {}
        file_as_df = pd.read_csv(os.path.join(source_path, df_file))
        by_file_folder = file_as_df.groupby('檔案資料夾')
        for folder_name, folder_frame in by_file_folder:
            by_path = folder_frame.groupby('路徑')
            counts[folder_name] = by_path.count()
            folder_names.append(folder_name)
            compare_box = counter = compare_name = length = flag = None
            for path_name, path_frame in by_path:
                print('.', end='')
                path_names.append(path_name)
                path_frames[path_name] = path_frame.sort_values(['路徑', '檔案名稱'])[['組別', '檔案類型', '檔案資料夾', '檔案名稱', '大小', '修改日期']].reset_index(drop=True)
                if compare_box is None:
                    compare_box = path_frames[path_name]
                    compare_name = path_name
                    counter = 0
                    continue
                counter+=1
                try:
                    if (compare_box == path_frames[path_name]).any(axis=None):
                        csv_writer.writerow([folder_name, compare_name, path_name, counter, len(compare_box), len(path_frames[path_name]), '相同', ''])
                    else:
                        csv_writer.writerow([folder_name, compare_name, path_name, counter, len(compare_box), len(path_frames[path_name]), '相異', ''])
                except Exception as e:
                    pointer += 1
                    csv_writer.writerow([folder_name, compare_name, path_name, counter, len(compare_box), len(path_frames[path_name]), pointer, e])
                    difference = compare_box.merge(path_frames[path_name], on='檔案名稱', how='outer', suffixes=['', '_'], indicator=True).loc[lambda x : x['_merge']!='both']
                    difference.set_index('檔案名稱', drop=True)
                    difference.to_csv(f'E:/.developing_env/duplicated_files/compare/{pointer}.csv', encoding='utf-8-sig')
                    results = pd.concat([results, difference])
        print()
results.to_csv('E:/.developing_env/duplicated_files/result.csv', encoding='utf-8-sig')
```
